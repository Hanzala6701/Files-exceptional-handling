1) What is a random variable in probability theory?

- A random variable in probability is a real-valued function that assigns a numerical value to each outcome of a random experiment (like flipping coins or rolling dice). It's called "random" because its value depends on chance, but it's "variable" because it can take different values for different outcomes, allowing us to analyze unpredictable events with math, typically represented by capital letters (X, Y).

2) What are the types of random variables?

- The main types of random variables are Discrete, which take countable, distinct values (like the number of heads in coin flips), and Continuous, which can take any value within a range (like height or temperature). A less common type is the Mixed random variable, which combines characteristics of both.

3) Explain the difference between discrete and continuous distributions.

- Discrete distributions involve countable, distinct outcomes (like rolling a die), while continuous distributions cover any value within a range (like height or temperature). The key difference is that discrete variables are counted, have gaps (e.g., 1, 2, 3), and have non-zero probabilities for single points, whereas continuous variables are measured, form a smooth continuum, and have zero probability for any exact single point, requiring probability density functions for ranges.

4) What is a binomial distribution, and how is it used in probability?

- A binomial distribution models the probability of getting a specific number of "successes" in a fixed number of independent trials, where each trial has only two outcomes (success or failure) with the same probability of success. It's used in probability to calculate chances for yes/no scenarios, like coin flips or product defects, by defining the number of trials (\(n\)) and the probability of success (\(p\)), helping analyze risk, quality, and statistical significance in fields from manufacturing to medicine.

5)  What is the standard normal distribution, and why is it important?

-  The standard normal distribution (Z-distribution) is a specific normal distribution with a mean of 0 and a standard deviation of 1, characterized by its bell-shaped curve, and it's crucial because it standardizes any normal distribution, allowing comparisons (via Z-scores) and probability calculations for diverse datasets like test scores or heights, using universal Z-tables, thanks to the Central Limit Theorem and Empirical Rule (68-95-99.7).

6) What is the Central Limit Theorem (CLT), and why is it critical in statistics?

- The Central Limit Theorem (CLT) states that the distribution of sample means from any population (regardless of its shape) will approximate a normal (bell-shaped) distribution as the sample size gets sufficiently large (often n>30), with the sample mean approaching the population mean. It's critical because it allows statisticians to make reliable inferences about large populations using small samples, forming the foundation for hypothesis testing, confidence intervals, and quality control, even when the original data isn't normal.

7) What is the significance of confidence intervals in statistical analysis?

- Confidence intervals (CIs) are crucial in statistics for quantifying the uncertainty around an estimate, providing a range where the true population parameter likely falls, not just a single point, which offers a more complete picture of reliability, effect size, and precision than p-values alone, guiding better evidence-based decisions in research, medicine, and business. They show how stable an estimate is: narrower CIs suggest greater precision and stability, while wider ones indicate more variability.

8) What is the concept of expected value in a probability distribution?

- The expected value (E(X) or Î¼) in a probability distribution is the theoretical average outcome of a random event if repeated infinitely, calculated by multiplying each possible outcome by its probability and summing those products, representing the long-run center of the distribution, often called the mean or first moment. It's a single value summarizing a distribution, indicating where results tend to cluster, but not necessarily an achievable result in a single trial.

```python
#9)Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

# --- 1. Generate 1000 random numbers from a normal distribution ---
# Parameters for the normal distribution
mean = 50
std_dev = 5
num_samples = 1000

# Generate the random numbers using NumPy's random.normal function
# Use a fixed seed for reproducibility if needed (optional)
np.random.seed(42)
random_numbers = np.random.normal(mean, std_dev, num_samples)

print(f"Generated {len(random_numbers)} random numbers.\n")

# --- 2. Compute mean and standard deviation using NumPy ---
computed_mean = np.mean(random_numbers)
computed_std_dev = np.std(random_numbers)

print(f"Computed Mean: {computed_mean:.4f}")
# Expected mean is close to 50
print(f"Computed Standard Deviation: {computed_std_dev:.4f}\n")
# Expected std dev is close to 5

# --- 3. Draw a histogram to visualize the distribution ---
plt.figure(figsize=(8, 6))

# Plot the histogram
plt.hist(random_numbers, bins=30, density=False, alpha=0.7, color='blue', edgecolor='black')

# Add titles and labels
plt.title(f'Histogram of Normally Distributed Random Numbers (Mean={mean}, Std Dev={std_dev})')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

# Add vertical lines for the theoretical mean and the computed mean
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Theoretical Mean ({mean})')
plt.axvline(computed_mean, color='green', linestyle='dashed', linewidth=2, label=f'Computed Mean ({computed_mean:.2f})')
plt.legend()

# Display the plot
plt.show()
```
