1) What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample data to draw conclusions about a population. It involves formulating two competing hypotheses, the null hypothesis (H0) and the alternative hypothesis (Ha), and then collecting data to assess the evidence.

2) What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis is a statement of "no effect" or "no difference," while the alternative hypothesis is a statement that contradicts the null hypothesis, suggesting there is a significant effect or difference. Essentially, the null hypothesis assumes any observed differences are due to random chance, whereas the alternative hypothesis suggests the differences are due to a real effect.

3)Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
-  Hypothesis Testing and Confidence Intervals | AnalystPrep ...In hypothesis testing, the significance level (often denoted as alpha or α) is a probability value that defines the threshold for determining whether the results of a statistical test are statistically significant. It represents the probability of rejecting the null hypothesis when it is actually true (a Type I error). Essentially, it sets the benchmark for how unusual a result must be (given the null hypothesis is true) before we're willing to reject that hypothesis.

4) What are Type I and Type II errors? Give examples of each.
- In hypothesis testing, Type I and Type II errors represent two ways a conclusion can be incorrect. A Type I error (false positive) occurs when a true null hypothesis is rejected, while a Type II error (false negative) occurs when a false null hypothesis is not rejected. Essentially, Type I errors involve claiming an effect when there isn't one, and Type II errors involve missing an existing effect.

```python
# 7) Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
# Code -
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis test parameters
mu_0 = 50       # Population mean under H0
sigma = 1       # Assumed population standard deviation
alpha = 0.05    # Significance level

# Sample statistics
n = len(sample_data)
sample_mean = np.mean(sample_data)

# Z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject H0: There is significant evidence that the population mean ≠ 50.")
else:
    print("Fail to Reject H0: No significant evidence that the population mean differs from 50.")
```

```python
# 8) Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

# - The following Python script simulates data from a normal distribution, calculates the 95% confidence interval for its mean, and visualizes the data and the confidence interval using Matplotlib.
# Code -
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Simulate data from a normal distribution
mu_true = 50  # True mean of the population
sigma_true = 5  # True standard deviation of the population
sample_size = 100 # Number of data points to simulate

np.random.seed(42) # for reproducibility
data = np.random.normal(loc=mu_true, scale=sigma_true, size=sample_size)

# 2. Calculate the 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1) # Use ddof=1 for sample standard deviation
standard_error = sample_std / np.sqrt(sample_size)

confidence_level = 0.95
degrees_freedom = sample_size - 1
alpha = 1 - confidence_level

# Calculate the t-critical value
t_critical = stats.t.ppf(1 - alpha / 2, df=degrees_freedom)

# Calculate the margin of error
margin_of_error = t_critical * standard_error

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({lower_bound:.2f}, {upper_bound:.2f})")

# 3. Plot the data and the confidence interval
plt.figure(figsize=(10, 6))
plt.hist(data, bins=10, density=True, alpha=0.6, color='skyblue', label='Simulated Data Histogram')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Sample Mean: {sample_mean:.2f}')
plt.axvspan(lower_bound, upper_bound, color='green', alpha=0.3, label='95% Confidence Interval')
plt.title('Simulated Normal Distribution Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 9) Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

# - A Python function can be created to calculate Z-scores and visualize standardized data.
# Code -
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def analyze_data_with_zscore(data):
    """
    Calculates Z-scores for a dataset and visualizes the standardized data
    using a histogram.

    Args:
        data (array-like): The input dataset.
    """
    # Calculate Z-scores
    z_scores = zscore(data)
    print(f"Original Data: {data}")
    print(f"Calculated Z-Scores: {z_scores}")

    # Visualize standardized data
    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=10, edgecolor='black')
    plt.title('Histogram of Standardized Data (Z-Scores)')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

# Example usage:
sample_data = np.array([10, 12, 15, 18, 20, 22, 25, 28, 30, 32])
analyze_data_with_zscore(sample_data)
```
